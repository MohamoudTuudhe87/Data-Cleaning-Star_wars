```python
favorably_counts = {}
unfavorably_counts = {}
controversial_counts = {}


for col in star_wars.columns[15:28]:
    counts = star_wars[col].value_counts()
    
    favorably_counts[col] = counts.get('Favorably', 0)
    unfavorably_counts[col] = counts.get('Unfavorably', 0)

characters_score = {characters :abs(favorably_counts[characters] - unfavorably_counts[characters])
                                    for characters in favorably_counts.keys()}

Most_favorably = max(favorably_counts, key=favorably_counts.get)
Most_unfavorably = max(unfavorably_counts, key=unfavorably_counts.get)
Most_controversial = min(characters_score, key=characters_score.get)

print(f"The Most Liked Character is : {Most_favorably}")
print(f"The Most Unliked Character is : {Most_unfavorably}")
print(f"The Most Most controversial Character is : {Most_controversial}")
```

```python
characters = list(favorably_counts.keys())
Most_favorably_values = list(favorably_counts.values())
Most_Unfavorably_values = list(unfavorably_counts.values())

plt.figure(figsize =(10,5))
plt.barh(characters, Most_favorably_values, color='green', label='Favorably')
plt.barh(characters, Most_Unfavorably_values, color='red', label='Unfavorably', left=Most_favorably_values)

plt.xlabel('Counts')
plt.ylabel('Characters')
plt.title('Most Liked Character vs Most Unliked Character')
plt.legend()

plt.tight_layout()
plt.show()
```
